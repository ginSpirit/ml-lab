````python
#Unit sample[Impulse] signal
n = np.arange(-3, 4, 1)
delta_n = np.zeros(len(n))
delta_n[n == 0] = 1

print(delta_n)        # [0. 0. 0. 1. 0. 0. 0.]

plt.stem(n, delta_n)
plt.title('Unit Sample (Impulse) Signal')
plt.xlabel('n')
plt.ylabel('Î´[n]')
plt.show()
````

````python
#Unit Ramp Signal
n = np.arange(-3, 4, 1)
x_n = np.zeros(len(n))
x_n[n >= 0] = n[n >= 0]

print(x_n)

plt.stem(n, x_n)
plt.title('Unit Ramp Signal')
plt.xlabel('n')
plt.ylabel('r[n]')
plt.show()
````

````python
#Unit Step Signal
n = np.arange(-3, 4, 1)
u_n = np.zeros(len(n))
u_n[n >= 0] = 1

print(u_n)

plt.stem(n, u_n)
plt.title('Unit Step Signal')
plt.xlabel('n')
plt.ylabel('u[n]')
plt.show()
````

````python
# Scaling

x_n = np.array([-2, 1, 0, 7, 2, 3, 1, 6, -1])
n = np.arange(-3, 6, 1)
a = 2
````

````python
#2n
idx = [i for i, val in enumerate(n) if val % a == 0]
x_scaled = x_n[idx]
n_scaled = n[idx] // a

plt.stem(n_scaled, x_scaled)
plt.title('Time-Scaled Signal: x[2n]')
plt.xlabel('n')
plt.ylabel('x[2n]')
plt.show()
````

````python
#n/2
length = a * (len(x_n) - 1) + 1
new_x_n = np.zeros(length, dtype=int)
n_new = np.arange(n[0] * a, n[-1] * a + 1)
for i in range(len(x_n)):
    if 0 <= i * a < length:
        new_x_n[i * a] = x_n[i]

n_new = np.arange(n[0] * a, n[0] * a + length)


plt.stem(n_new, new_x_n)
plt.title('Time-Scaled Signal: x[n/2]')
plt.xlabel('n')
plt.ylabel('x[n/2]')
plt.show()
````
