# task-2: take a Gray Scale image of an umbrella and convert it into binary with thresholds(80,150,200)




```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_path = '/content/drive/MyDrive/img/umbrella.jpg'
```

```python
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Display RGB Image
plt.figure(figsize=(5,5))
plt.imshow(image_rgb)
plt.title('RGB Image')
plt.axis('off')
plt.show()
```

```python
# Diaplay in grayscale image
plt.figure(figsize=(5,5))
plt.imshow(image_gray, cmap='gray', vmin=0, vmax=255)
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
image_bi = np.zeros_like(image_gray)
rows, cols = image_gray.shape
for i in range(rows):
    for j in range(cols):
      if image_gray[i,j] < 80:
        image_bi[i,j] = 0
      else :
        image_bi[i,j] = 255
```

```python
plt.title('split by 80')
plt.imshow(image_bi, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.title('split by 150')
plt.imshow(image_bi, cmap='gray')
plt.axis('off')
plt.show()
```
