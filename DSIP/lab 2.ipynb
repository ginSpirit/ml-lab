```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(30)
a = -0.85
plt.stem(n, a**n)
plt.title('Signal for (–1 < a < 0)')
plt.xlabel('n'); plt.ylabel('x(n)')
plt.show()
```
