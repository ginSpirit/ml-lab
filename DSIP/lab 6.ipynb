```python
import numpy as np
import matplotlib.pyplot as plt

# Input sequence and index values
x = np.array([1, 2, -1, 4, 3, 2])
n = np.arange(-3, 3)

# Print Z-transform expression
if len(x) > 0:
    print(f"{x[0]}*z^(-{n[0]})", end="")

for i in range(1, len(x)):
    print(f" + {x[i]}*z^(-{n[i]})", end="")
print()  # new line after printing

# Plot the discrete sequence
plt.stem(n, x, basefmt=" ")
plt.title("Discrete Signal for Z-Transform")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.show()
```
