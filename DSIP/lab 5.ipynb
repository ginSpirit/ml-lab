```python
import numpy as np
import matplotlib.pyplot as plt

x_n = np.array([-1, 2, 3, 0, 1, 5, -2, 3, 1])
n = np.arange(-2, 7)

k = int(input("Enter shift amount k: "))

n_shifted = n + k
start = min(n.min(), n_shifted.min())
stop = max(n.max(), n_shifted.max())
nt = np.arange(start, stop+1)

x_plot = np.zeros(len(nt), int)
idx = n_shifted - start
x_plot[idx] = x_n

plt.figure(figsize=(8,4))
plt.stem(nt, x_plot, basefmt=" ")
plt.axhline(y=0, color='red', linewidth=1)
plt.xlabel("n")
plt.ylabel("x(n)")
title = "Delayed" if k>0 else ("Original" if k==0 else "Advanced")
plt.title(f"{title} Signal (k={k})")
plt.grid(True)
plt.show()
```
