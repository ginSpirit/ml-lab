```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,2,0.001)
x_t = np.sin(2*np.pi*5*t)
plt.plot(t,x_t)
plt.title('Sin Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,2,0.001)
x_t = np.cos(2*np.pi*5*t)
plt.plot(t,x_t)
plt.title('Cos Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 2000)
plt.plot(t, np.cos(2*np.pi*5*t),
         t, np.sin(2*np.pi*5*t))
plt.title('Cosine and Sine Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 2000)
signals = [np.sin(2*np.pi*5*t), -np.sin(2*np.pi*5*t)]

fig, ax = plt.subplots()
for s in signals:
    ax.plot(t, s)
ax.set(title='Sine & Negative Sine', xlabel='Time', ylabel='Amplitude')
plt.show()
```
