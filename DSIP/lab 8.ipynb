```python
image_path = '/content/drive/MyDrive/img/test.jpg'
```

```python
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Display Image
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.title('RGB Image')
plt.axis('off')
plt.show()
```

```python
# Diaplay in grayscale image
plt.figure(figsize=(6,6))
plt.imshow(image_gray, cmap='gray', vmin=0, vmax=255)
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
image_negative = 255 - image_gray

plt.figure(figsize=(6,6))
plt.imshow(image_negative, cmap='gray', vmin=0, vmax=255)
plt.title('Negative Image')
plt.axis('off')
plt.show()
```

```python
def apply_gamma_gray(gray_img, gamma, c=1.0):
    arr = gray_img.astype(np.float32) / 255.0
    out = c * np.power(arr, float(gamma))
    out = np.clip(out, 0.0, 1.0)
    return (out * 255.0).astype(np.uint8)
```

```python
# Gamma < 1 (Bright)

gamma1 = 0.5
image_gamma1 = apply_gamma_gray(image_gray, gamma=gamma1, c=1.0)

plt.figure(figsize=(6,6))
plt.imshow(image_gamma1, cmap='gray', vmin=0, vmax=255)
plt.title(f'Gamma Transformation ( = {gamma1} < 1)  Brighter')
plt.axis('off')
plt.show()


```

```python

# Gamma > 1 (Dark)
gamma1 = 2.0
image_gamma1 = apply_gamma_gray(image_gray, gamma=gamma1, c=1.0)

plt.figure(figsize=(6,6))
plt.imshow(image_gamma1, cmap='gray', vmin=0, vmax=255)
plt.title(f'Gamma Transformation ( = {gamma1} > 1)  Brighter')
plt.axis('off')
plt.show()


```
