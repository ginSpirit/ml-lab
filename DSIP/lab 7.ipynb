```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import pandas as pd
import cv2
```

```python
file_path = '/content/drive/MyDrive/img/ponyo.jpg'
```

```python
image = cv2.imread(file_path)
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#display image
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
binary_img = np.zeros_like(image_gray)
row, colmn = image_gray.shape
for i in range(row):
  for j in range(colmn):
    if image_gray[i, j] <= 128:
      binary_img[i, j] = 0
    else:
      binary_img[i, j] = 255
```

```python
plt.imshow(binary_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
file_path_of_darkimg = '/content/drive/MyDrive/img/spirit.jpg'
```

```python
dark_image = cv2.imread(file_path_of_darkimg)
img_rgb = cv2.cvtColor(dark_image, cv2.COLOR_BGR2RGB)

#display image
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
image_gray = cv2.cvtColor(dark_image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.hist(cv2.cvtColor(dark_image, cv2.COLOR_BGR2GRAY).ravel(), bins=256)
plt.title("Histogram of Dark Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Count")
plt.xlim([0, 255])
plt.show()
```
